```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.linalg import lstsq # for the multiple linear regression, we'll use the scipy linear algebra least-squares function
from scipy.interpolate import interp1d # for quantile regression, we'll want this 1d interpolation function
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in a .csv file
data = pd.read_excel('E:\\cascades_swe.xlsx')
data.head()
```

```python
y = data['April_1_SWE']
c = ['oct_mar_precip','oct_mar_tmax']
x = data[c]
```

```python
# use the linear regression function
B1, B0, rvalue, pvalue, sB1 = stats.linregress(data.oct_mar_precip, data.April_1_SWE)
```

```python
data['SWE_prec'] = B1 * data['oct_mar_precip'] + B0
```

```python
B1, B0, rvalue, pvalue, sB1 = stats.linregress(data.oct_mar_tmax, data.April_1_SWE)
```

```python
data['SWE_tmax'] = B1 * data['oct_mar_tmax'] + B0
```

```python
B1, B0, rvalue, pvalue, sB1 = stats.linregress(data.oct_mar_tmin, data.April_1_SWE)
```

```python
data['SWE_tmin'] = B1 * data['oct_mar_tmin'] + B0
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()# 定义模型
model.fit(x,y)
```

```python
print("参数w={},b={}".format(model.coef_,model.intercept_))
```

```python
B1=model.coef_[0]
B2 = model.coef_[1]
B0 = model.intercept_
```

```python
data['SWE_mul'] = B1 * data['oct_mar_precip'] + B2 * data['oct_mar_tmax'] + B0
```

```python
fig, ax = plt.subplots(figsize=(10,5))

data.plot(x='Water_Year', y='April_1_SWE', c='r', linestyle='-', linewidth=1, marker='x', ax=ax, label='April 1 sWE Observed')
data.plot(x='Water_Year', y='SWE_prec', c='y', linestyle='-', linewidth=1, marker='.', ax=ax, label='pre_swE_prec with Simple Linear Regression')
data.plot(x='Water_Year', y='SWE_tmax', c='c', linestyle='-', linewidth=1, marker='.', ax=ax, label='pre_sWE_tmax with Simple Linear Regression')
data.plot(x='Water_Year', y='SWE_tmin', c='m', linestyle='-', linewidth=1, marker='.', ax=ax, label='pre_sWE_tmin with Simple Linear Regression')
data.plot(x='Water_Year', y='SWE_mul', c='black', linestyle='--', linewidth=1, marker='.', ax=ax, label='April 1 sWE Predicted with Multiple Linear Regression')
ax.set_title('Timeline of Peak Snow Water Equivalent (SWE)', fontsize=15)
ax.set_xlabel('Water Year', fontsize=12)
ax.set_ylabel('Peak SWE (mm)', fontsize=12);
plt.legend(loc=(0.2,-0.36), fontsize=12);
```
